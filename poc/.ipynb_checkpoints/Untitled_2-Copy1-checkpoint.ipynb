```python
import tabula
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
stations = pd.read_excel("mtr_station.xlsx")
stations = stations["station"].unique()
```

```python
file = "ctn_20220127.pdf"

pages = tabula.read_pdf_with_template(f"cache/{file}", template_path="t2.json", lattice=True, pandas_options={'header': None})

df = pd.DataFrame()
for page in pages:
    page = page.astype(str)
    dc = [page[col].str.contains("Specified place").any() for col in page.columns]
    df = pd.concat([df, page.loc[:, dc]], ignore_index=True).astype(str)
# df
```

```python
df_merge = pd.DataFrame()

for col in df.columns:
    df_merge = pd.concat([df_merge, df.fillna("na")[col].str.contains(r"[^\x00-\x7F]", regex=True)], axis=1)
# df_merge
```

```python
file = "ctn_20220521.pdf"

pages = tabula.read_pdf_with_template(f"cache/{file}", template_path="config/ctn_template.json", lattice=True)

df = pd.DataFrame()
for page in [page for page in pages if page.dropna(how="all").shape[0]]:
    df_merge = page.dropna(how="all") \
        .apply(lambda y: "".join(y.dropna().astype(str)), axis=1) \
        .str.replace(r"\r|\s|’|é|及|或|指明地點|視情況而|年年年年", "", regex=True) \
        .str.extract(r"([^\x00-\x7F]+[\s\d\w\(\)-]{0,}[^\x00-\x7F]+[\s\d\w\(\)-]{0,}[^\x00-\x7F]+[\)]{0,})") \
        .dropna() \
        .set_axis(["specified_place"], axis=1)
    df = pd.concat([df, df_merge], ignore_index=True)

df
```

```python
file = "ctn_20220115.pdf"

pages = tabula.read_pdf_with_template(f"cache/{file}", template_path="ctn_template.json", lattice=True)

df = pd.DataFrame()
for page in [page for page in pages if page.dropna(how="all").shape[0]]:
    df_merge = page.dropna(how="all") \
        .apply(lambda y: "".join(y.dropna().astype(str)), axis=1) \
        .str.replace(r"\r|\s|’|é|及|或|指明地點", "", regex=True) \
        .str.extract(r"([^\x00-\x7F]+[\s\d\w\(\)-]{0,}[^\x00-\x7F]+[\s\d\w\(\)-]{0,}[^\x00-\x7F]+[\)]{0,})") \
        .dropna() \
        .set_axis(["specified_place"], axis=1)
    df = pd.concat([df, df_merge], ignore_index=True)

df
```

```python
def remove_unrelated_words(wds):
    
    return ''.join([wd for wd in wds if "曾身處指明場" not in wd])

df = df["table_str"].str.extract(r"(\d+\..*[^\x00-\x7F]+)") \
    .dropna() \
    .reset_index(drop=True) \
    .rename(columns={0: "specified_place"})

df = df["specified_place"].str.replace(r"\r|或|及", "", regex=True) \
    .str.findall(r"([^\x00-\x7F]+[\s\d\w\(\)]{0,}[^\x00-\x7F]+[\s\d\w\(\)]{0,}[^\x00-\x7F]+)") \
    .to_frame()

df["specified_place"] = df["specified_place"].map(remove_unrelated_words)

df.head()
```
